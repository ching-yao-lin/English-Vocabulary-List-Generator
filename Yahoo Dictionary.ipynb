```python
vocabulary_list = []
with open('vocabulary.txt', 'r') as f:
    contents = f.readlines()
    for line in contents:
        vocabulary_list.append(line.strip())

with open('vocabulary list.txt', 'w') as f:
    pass
```

```python
class YahooDictionarySpider(scrapy.Spider):
    """
    Items Inherited from scrapy: 
        name: the name of the Spider
        start_urls: a list of pages to start crawling with
        parse(self, response): a generator that has to yield an scrapy item (or a dict)
      
    Extracted Data -> Temporary Containers (items) -> store in database (or somewhere else)
    """
    
    name = 'yahoo' 
    start_urls = ['https://tw.dictionary.search.yahoo.com/search?p='+word+'&ei=UTF-8&fr=sfp' for word in vocabulary_list]
    
    def parse(self, response):
        
#         items = YahooDictionaryItem()
        
        vocab = response.css(".lh-24::text").extract_first() # 單字
#         part_of_speech = response.css(".mb-12 .tabs-pos_type::text").extract() # 詞性
#         meaning = response.css(".mb-12 > span:nth-child(1)::text").extract() # 意思
        contents = response.css(".ml-50.mb-12 .fz-14").extract()
        
        contents = [item.replace('<b>', '').replace('</b>', '') for item in contents]
        
        pattern = re.compile(r'<[a-zA-Z0-9 _=/"-]+>')
        tags = [pattern.findall(item) for item in contents]
        
        for tag in tags:
            contents = [item.replace(tag[0], '').replace(tag[1], '') for item in contents]
        
#         items['vocab'] = vocab
#         items['part_of_speech'] = part_of_speech
#         items['meaning'] = meaning
        
        part_of_speeches = ['vt.', 'vi.', 'adj.', 'adv.', 'n.', 'n.[C]', 'n.[U]', 'prep.', 'conj.', 'abbr.', 'ph.']
        with open('vocabulary list.txt', 'a') as f:
            f.write(vocab + '\n')
            for content in contents:
                if content in part_of_speeches:
                    f.write(content + ' ')
                else:
                    f.write(content + '\n')
            f.seek(-1, os.SEEK_CUR)
        
#         print(vocab)
#         print(part_of_speech)


#         yield items
        
```

```python
part_of_speeches = ['vt.', 'vi.', 'adj.', 'adv.', 'n.', 'n.[C]', 'n.[U]', 'prep.', 'conj.', 'abbr.', 'ph.']
def isPOS(string):
    for elmnt in part_of_speeches:
        if string.startswith(elmnt):
            return True
    return False

count_lines = 0
with open('vocabulary list.txt', 'r', newline='') as f:    
    for line in f:
        if line.strip().isalpha(): # a vocabulary
            print('VOCAB:', line, end='')
        elif isPOS(line): # is part of speech
            print('POS:', line, end='')
        else: # meaning and sentences
            count_lines += 1
            print(line, end='')
```

```python
import openpyxl
from openpyxl.styles import Font

wb = openpyxl.Workbook()
sheet = wb.active
sheet.title = 'My Vocabs'

with open('vocabulary list.txt', 'r', newline='') as f:
    i = 1
    for line in f:
        print(line, end='')
        if line.strip().isalpha(): # a vocabulary
            sheet['A' + str(i)] = line.strip()
            sheet['A' + str(i)].font = Font(bold=True)
        elif isPOS(line): # part of speech
            sheet['B' + str(i)] = line.strip()
        else: # meaning and sentences
            sheet['C' + str(i)] = line.strip()
            i += 1

wb.save("vocabulary words.xlsx")
```
